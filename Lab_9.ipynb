# Shortest Path Using Dijkstra Algorithm in Python
Problem Description:
Given a weighted graph with non negative edge weights. Find the shortest distance from a source vertex to all other vertices using Dijkstra algorithm

#Theory
Dijkstra algorithm finds shortest distance from one source node to all other nodes in a weighted graph with non negative weights. Dictionary structure stores nodes as keys and neighbor nodes with weights as values.

##Algorithm Steps

1. Represent graph using dictionary.
2. Set distance of every node as infinity.
3. Set distance of source node as 0.
4. Select node with smallest distance among unvisited nodes.
5. Update distance of neighbor nodes.
6. Repeat until all nodes get visited.

```python
def dijkstra(graph, start):

    distances = {}
    predecessors = {}
    visited = []

    for node in graph:
        distances[node] = float('inf')
        predecessors[node] = None

    distances[start] = 0

    while len(visited) < len(graph):

        min_node = None
        min_distance = float('inf')

        for node in graph:
            if node not in visited and distances[node] < min_distance:
                min_distance = distances[node]
                min_node = node

        if min_node is None:
            break

        visited.append(min_node)

        for neighbor, weight in graph[min_node].items():
            new_distance = distances[min_node] + weight

            if new_distance < distances[neighbor]:
                distances[neighbor] = new_distance
                predecessors[neighbor] = min_node

    return distances, predecessors
```

```python
def reconstruct_path(predecessors, start, end):
    path = []
    current = end
    while current is not None:
        path.insert(0, current)
        current = predecessors[current]

    if path[0] != start:
        return []
    return path
```

```python
import heapq

def dijkstra(graph, start):
    distances = {node: float('inf') for node in graph}
    distances[start] = 0
    pq = [(0, start)]

    while pq:
        current_distance, current_node = heapq.heappop(pq)

        if current_distance > distances[current_node]:
            continue

        for neighbor, weight in graph[current_node].items():
            distance = current_distance + weight

            if distance < distances[neighbor]:
                distances[neighbor] = distance
                heapq.heappush(pq, (distance, neighbor))

    return distances
```

```python

graph_updated = {
    'A': {'B': 4, 'C': 2},
    'B': {'A': 4, 'C': 5, 'D': 10, 'E': 7},
    'C': {'A': 2, 'B': 5, 'D': 3, 'F': 8},
    'D': {'B': 10, 'C': 3, 'E': 4, 'F': 6},
    'E': {'B': 7, 'D': 4, 'F': 3},
    'F': {'C': 8, 'D': 6, 'E': 3}
}


print("\n--- Testing with source 'A' ---")
distances_A, predecessors_A = dijkstra(graph_updated, 'A')

for node in distances_A:
    path = reconstruct_path(predecessors_A, 'A', node)
    if path:
        print(f"Shortest distance from A to {node} = {distances_A[node]}, Path: {' -> '.join(path)}")
    else:
        print(f"No path from A to {node}")


print("\n--- Testing with source 'B' ---")
distances_B, predecessors_B = dijkstra(graph_updated, 'B')

for node in distances_B:
    path = reconstruct_path(predecessors_B, 'B', node)
    if path:
        print(f"Shortest distance from B to {node} = {distances_B[node]}, Path: {' -> '.join(path)}")
    else:
        print(f"No path from B to {node}")
```

```python
def get_user_graph():
    print("Enter your graph in dictionary format. Example: {'A': {'B': 1, 'C': 4}, 'B': {'A': 1, 'C': 2}, 'C': {'A': 4, 'B': 2}}")
    while True:
        try:
            user_input = input("Enter graph: ")
            user_graph = eval(user_input)

            if not isinstance(user_graph, dict) or not all(isinstance(v, dict) for v in user_graph.values()):
                raise ValueError("Invalid graph format. Please ensure it's a dictionary of dictionaries.")
            return user_graph
        except (SyntaxError, NameError, ValueError) as e:
            print(f"Invalid input or format error: {e}. Please try again.")

def get_valid_node(prompt, graph_nodes):
    while True:
        node = input(prompt).strip()
        if node in graph_nodes:
            return node
        else:
            print(f"Invalid node. Please choose from: {', '.join(graph_nodes)}")
user_graph = get_user_graph()


if user_graph:
    source_node = get_valid_node("Enter the source node: ", user_graph.keys())

    distances_user, predecessors_user = dijkstra(user_graph, source_node)

    print(f"\n--- Shortest Paths from Source '{source_node}' ---")
    for node in distances_user:
        path = reconstruct_path(predecessors_user, source_node, node)
        if path:
            print(f"Shortest distance to {node} = {distances_user[node]}, Path: {' -> '.join(path)}")
        else:
            print(f"No path from {source_node} to {node}")
else:
    print("No graph entered or graph was empty. Cannot run Dijkstra.")
```
