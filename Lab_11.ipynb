# Travelling Salesman Problem using Dynamic Programming.
Problem Description:
Given a set of cities and distances, find the shortest possible route that visits each city once and returns to the starting city.

#Theory
TSP uses dynamic programming with bit masking. Store results of visited subsets to reduce repeated work.

## Algorithm
1. Start from source city
2. Mark visited cities using bit mask
3. Try all unvisited cities
4. Choose minimum cost path
5. Return to source city

```python
def tsp(graph):
    n = len(graph)
    dp = [[-1]*(1<<n) for _ in range(n)]

    path_memo = [[-1]*(1<<n) for _ in range(n)]

    def visit(city, mask):
        if mask == (1<<n) - 1:
            return graph[city][0]

        if dp[city][mask] != -1:
            return dp[city][mask]

        ans = float('inf')
        best_next_city = -1

        for next_city in range(n):
            if (mask >> next_city) & 1 == 0:
                new_cost = graph[city][next_city] + visit(next_city, mask | (1<<next_city))
                if new_cost < ans:
                    ans = new_cost
                    best_next_city = next_city

        dp[city][mask] = ans
        path_memo[city][mask] = best_next_city
        return ans

    min_cost = visit(0, 1)
    path = []
    current_city = 0
    current_mask = 1
    path.append(current_city)

    while len(path) < n:
        next_c = path_memo[current_city][current_mask]
        if next_c == -1:
            break
        path.append(next_c)
        current_mask |= (1 << next_c)
        current_city = next_c

    path.append(0)

    return min_cost, path
```

```python
graph = [
[0, 10, 15, 20],
[10, 0, 35, 25],
[15, 35, 0, 30],
[20, 25, 30, 0]
]

min_cost, path = tsp(graph)
print("Minimum travelling cost:", min_cost)
print("Shortest path:", path)
```

```python
def get_user_graph():
    while True:
        try:
            n = int(input("Enter the number of cities (e.g., 4): "))
            if n <= 1:
                print("Number of cities must be at least 2.")
                continue
            break
        except ValueError:
            print("Invalid input. Please enter an integer.")

    print(f"""Enter the {n}x{n} adjacency matrix row by row. Use spaces to separate distances.
(e.g., for 4 cities, first row might be: 0 10 15 20)""")

    graph = []
    for i in range(n):
        while True:
            try:
                row_str = input(f"Enter row {i} (space-separated distances): ")
                row = list(map(int, row_str.split()))
                if len(row) != n:
                    print(f"Row must contain exactly {n} distances. Please try again.")
                    continue
                graph.append(row)
                break
            except ValueError:
                print("Invalid input. Please enter integers separated by spaces.")
    return graph

user_graph = get_user_graph()
print("Your entered graph:", user_graph)
```

```python
min_cost_user, path_user = tsp(user_graph)
print("Minimum travelling cost for user graph:", min_cost_user)
print("Shortest path for user graph:", path_user)
```
